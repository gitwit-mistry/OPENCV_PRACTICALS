## ðŸ“Œ Take 2 image crop some part of both image and swap it. 

```python
image1 = cv2.imread('riri.jpg')
image1 = cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
image2 = cv2.imread('JR.png')
image2 = cv2.cvtColor(image2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image1)
```

```python
plt.imshow(image2)
```

```python
image1_crop = image1[10:250,100:350,:].copy()
plt.imshow(image1_crop)
```

```python
offset = 100
image2_crop = image2[10+offset:250+offset,100+offset:350+offset,:].copy()
plt.imshow(image2_crop)
```

```python
image1[10:250,100:350,:] = image2_crop
```

```python
plt.imshow(image1)
```

```python
offset = 100
image2[10+offset:250+offset,100+offset:350+offset,:] = image1_crop
```

```python
plt.imshow(image2)
```
